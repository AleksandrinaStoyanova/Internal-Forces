# Internal Forces

### Introduction

The determination of the $\mathbf{Internal\ Forces}$ is one of the most important tasks in structural section sizing (it's extremum values).
The folowing conclusions and calculations are made presumpting the material is homogeneous, meaning it has identical properties at every point.
This is a major presumpion in the teorhy of Strength of Materials.
Used sources - lectures and exercise "Strength of Materials" for civil engineering in Univercity of Arhitecture, Civil Engineering and Geodesy, lecturer - Prof. Alexandrov.
The considered calculations are made for Statically determinated system (the number of unknowns are equal to the number of independent equilibrium equations) and static external forces (constant loads, that do not change in magnitude, direction or position over time).

Let's consider a beam, loaded with external forces $F_i$ thet is balanced (fig. 1)





```python
Image(filename="Fig_1.jpg", width=300, height=200)
```

We determine the support reactions of the beam using static equilibrium 
(the sum of external forces $\sum{F_x} = 0,\  \sum{F_y} = 0$ and moments $\sum{M} = 0$ equals zero, holding the structure stationary).
Let's cut the beam at some place, perpendicular to the axis of the beam with a plane $\alpha$. We 'll call the two parts $\mathbf{left}$ and $\mathbf{right}$. The plane $\alpha$ divides the forces into two components: $\vec{F_L}$ and $\vec{F_R}$. $\vec{F_L}$ includes the forces, acting on the left part of the beam and $\vec{F_R}$ - the forces, acting on the right part. At the point where we have cut the beam,there are an infinite number of internal forces, randomly orientated $S_i$, with witch the left and the right part of the beam react with each other. $S_L^{R}$ are the internal forces from the right part of the beam, acting on the left part, $S_R^{L}$ are the internal forces from the left part of the beam, acting on the right part.
Acording to Newton's Third Law ("For every action there is an equal and opposite reaction") - $S_L^{R},\ S_R^{L}\sim 0$.
Let's concider the left part of the beam. As we assumed that the beam is in balance, than every part of it will also be balanced. So we can write that $F_L,\ S_L^{R}\sim 0$.
The unknowns from this system are the internal forces $S_L^{R}$. They are an important characteristic for the state of the beam with the given external forces. As the internal forses $S_L^{R}$ are an infinite, randomly orientaded number, we can't calculate them from the equation $F_L,\ S_L^R\sim 0$, but we can calculate their resultant vector $\vec{R}$ and resultant moment $\vec{M}$ for the center of the masses of the cross section.
Then we can write $\vec{R} = - \vec{R_L}\\,\  \vec{M} = - \vec{M_L}$ for the left part of the beam and $\vec{R} = \vec{R_R},\ \vec{M} = \vec{M_R}$.
Combining the results from the two parts of the beam we come to:

$$(1)\          \left\{ \begin{array}{cc}\vec{R} = - \vec{R_L} = \vec{R_R} \\ 
\ \vec{M} = - \vec{M_L} = \vec{M_R}\end{array}\right.$$







```python
Image(filename="Fig_2.jpg", width=400, height=200)
```

If we project the resultant vector $\vec{R}$ and the resultant moment $\vec{M}$ on the normal axis of the beam and on the plane $\alpha$, which is perpendicular to the normal axis, we get the following forces and moments:\
$\vec{R} = \vec{N} + \vec{Q} = \vec{N} + \vec{Q_y} + \vec{Q_z} = N\vec{i} + Q_y\vec{j} + Q_z\vec{k}$\
$\vec{M} = \vec{M_t} + \vec{M_f} = \vec{M_t} + \vec{M_{f,y}} + \vec{M_{f,z}} = M_t\vec{i} + M_y\vec{j} + M_z\vec{k}$\
where $\vec{i},\ \vec{j},\ \vec{k}$ are the unit vectors along the axes $ x,\ y,\ z$\
$\vec{N}$ - normal force\
$\vec{Q}$ - tangential (shear) force\
$\vec{M_t}$ - torsional moment\
$\vec{M_f}$ - bending moment\

In summary $\vec{N},\ \vec{Q},\ \vec{M_t},\ \vec{M_f}$ are called $\mathbf{Internal\ Forces}$ in the plane $\alpha$ of the beam.


For practical purposes, it's easier to work with the scalar components of $\vec{R}$ and $\vec{M}$ than with the vectors themselfs.\
For this purpose we choose a special coordinate system, where the axis $\mathbf{x}$ matches with the external normal of the cross section\
and the axes $\mathbf{y}$ and $\mathbf{z}$ form a right - handed set of vectors.

$$(2)\begin{equation}
\left\{ 
  \begin{aligned}
    & N = R_x = - R_{L,x} = R_{R_x} - normal\ force \\
    & Q_y = R_y = - R_{L,y} = R_{R,y} - shear\ force\ over\ axis\ Cy \\
    & \ Q_z = R_z = - R_{L,z} = R_{R,z} - shear\ force\ over\ axis\ Cz \\
    & \ M_t = M_x = - M_{L,x} = M_{R,x} - torsional\ moment \\
    & \ M_{f,y} = M_y = - M_{L,y} = M_{R,y} - bending\ moment\ on\ Cy\ axis \\
    & \ M_{f,z} = M_z = - M_{L,z} = M_{R,z} - bending\ moment\ on\ Cz\ axis
  \end{aligned} 
\right.
\end{equation}$$

In reality it's quite common that the axis of the beam lies in the same plane as the external forces.\

If we take $xz$ to be that plane, than $Q_y = 0,\ M_t = 0,\ M_z = 0$.\

That kind of load is called $\mathbf{plane\ load}$.


### Calculating the Internal forces in a plane loaded beams

So if we want to determine the Internal forces in random cross section of a beam we follow this steps:
1. Calculate all reaction forces and moments at the supports using static equilibrium equations for the entire structure.
2. We make a cut at the point of interest and consider the left or the right part.
3. On the chosen part of the beam, we put all external forces and the internal forces, considered with positive directions.
4. We solve the three equlibrium equations to calculate the internal forces:
   
   $\sum{H} = 0\ -\ horizontal\ forces$
   
   
   $\sum{V} = 0\ -\ vertical\ forces$
   
   
   $\sum{M} = 0\ -\ moments\ about\ the\ cut$
   



For the design of a structure it's crutial to find the cross - section where the internal forces are with their extremum size - the maximum size of the internal forces is a measurement for the size of the internal forces.\
In practice it's best to show the internal forces like a diagram.

1. Defining the beam (it's geometry and loads):

```python
l = 4      # beam length
f = 40     # point load
a = 1      # distance from start to point load
q = 10     # distributed load
```

2. Drawing a simple beam with loads:

```python
def draw_simple_beam(l, f, a, q):
    plt.figure(figsize=(12,2))
    plt.plot([0, l], [0, 0], color="black", linewidth=5)
    plt.annotate('', xy=(a, 0), xytext=(a, 10),
                arrowprops=dict(facecolor='red', shrink=0.05, width=2))
    plt.text(a, 5, f'{f} kN', ha='right')
    plt.xlim(-1, l + 2)
    plt.ylim(-1, f)
    plt.axis('off')
    x_load = np.linspace(0, l, 100)
    plt.fill_between(x_load, 0, q, color='blue', alpha=0.3, label='UDL')
    plt.plot([-1, l], [q, q], color='blue', linewidth=2)
    arrow_x = np.linspace(0, l, 10)
    for x in arrow_x:
        plt.annotate('', xy=(x, 0), xytext=(x, q), 
                     arrowprops=dict(arrowstyle='->', color='blue', lw=1.5))
    plt.text(l/2, 5, f'{q} kN/m', ha='center')

    plt.ylim(-1, q )
    plt.xlim(0, l + 1)
    plt.title("Loads")
    plt.show()
```

```python
draw_simple_beam(l, f, a, q)
```

3. Calculating the reaction forces:

```python
def calculate_spring_reactions(l, f, a, q):
    reaction_1 = q * l / 2 + f * (l - a) / l
    reaction_2 = q * l / 2 + f * a / l
    return [reaction_1, reaction_2]
```

```python
calculate_spring_reactions(l, f, a, q)
```

4. Calculate the internal forces in different cross sections of the beam:


```python
def calculate_bending_moment(x, l, f, a, q):
    reaction_1, reaction_2 = calculate_spring_reactions(l, f, a, q)
    bending_moment = reaction_1 * x - (q * x * x) / 2 if x <= a else reaction_1 * x - (q * x * x) / 2 - f * (x - a)
    return bending_moment
```

```python
calculate_bending_moment(1, l, f, a, q)
```

Draw moment diagram and calculate bending moment extremum:

```python
x = np.linspace(0, l, 1000)

reaction_1 = calculate_spring_reactions(l, f, a, q)[0]
v_calculate_bending_moment = np.vectorize(calculate_bending_moment)
bending_moment = v_calculate_bending_moment(x, l, f, a, q)

max_moment = np.max(bending_moment)
max_idx_moment = np.argmax(bending_moment)
max_moment_location = x[max_idx_moment]

min_moment = np.min(bending_moment)
min_idx_moment = np.argmin(bending_moment)
min_moment_location = x[min_idx_moment]

plt.plot(x, bending_moment)
plt.fill_between(x, bending_moment, alpha=0.3)
plt.gca().invert_yaxis()
plt.scatter(max_moment_location, max_moment, color='red', label=f'Mmax={max_moment:.2f}')
plt.annotate(f"Mmax = {max_moment:.2f} kNm", 
             xy=(max_moment_location, max_moment), 
             xytext=(max_moment_location + 1, max_moment + 1))
plt.scatter(min_moment_location, min_moment, color='red', label='Mmin={min_moment:.2f}')
plt.annotate(f"Mmin = {min_moment:.2f} kNm", 
             xy=(min_moment_location, min_moment), 
             xytext=(min_moment_location + 0.2, min_moment + 2))

plt.title("Bending Moment Diagram")
plt.xlabel("L, m")
plt.ylabel("M, kN.m")
plt.grid(True)
plt.show()
```

```python
def calculate_shear_force(x, l, f, a, q):
    reaction_1, reaction_2 = calculate_spring_reactions(l, f, a, q)
    shear_force = reaction_1 - q * x  if x <= a else reaction_1 - q * x - f 
    return shear_force
```

```python
calculate_shear_force(1, l, f, a, q)
```

Draw shear diagram and calculate shear force extremum:

```python
x = np.linspace(0, l, 1000)

reaction_1 = calculate_spring_reactions(l, f, a, q)[0]
v_calculate_shear_force = np.vectorize(calculate_shear_force)
shear_force = v_calculate_shear_force(x, l, f, a, q)

max_shear_force = np.max(shear_force)
max_shear_idx = np.argmax(shear_force)
max_shear_location = x[max_shear_idx]

min_shear_force = np.min(shear_force)
min_shear_idx = np.argmin(shear_force)
min_shear_location = x[min_shear_idx]

shear_force_at_jump = v_calculate_shear_force(a, l, f, a, q)

plt.plot(x, shear_force)
plt.axhline(0, color='black', lw=1)
plt.fill_between(x, shear_force, alpha=0.3)
plt.gca()
plt.scatter(max_shear_location, max_shear_force, color='red')
plt.annotate(f"Qmax = {max_shear_force:.2f} kN", 
             xy=(max_shear_location, max_shear_force), 
             xytext=(max_shear_location + 0.3, max_shear_force - 1))
plt.scatter(min_shear_location, min_shear_force, color='red')
plt.annotate(f'Qmin = {min_shear_force:.2f} kN',
            xy=(min_shear_location, min_shear_force),
            xytext=(min_shear_location+0.3, min_shear_force + 1))
plt.scatter(a, shear_force_at_jump, color='red')
plt.annotate(f'Qjump = {shear_force_at_jump:.2f} kN',
            xy=(a, shear_force_at_jump),
            xytext=(a+0.2, shear_force_at_jump + 1))

plt.title("Shear Force Diagram")
plt.xlabel("L, m")
plt.ylabel("Q, kN")
plt.grid(True)
plt.show()
```

### Differential dependencies for the Internal forces

Let's concider a straight beam with random loads. We want to find the dependencies between  the Internal forces and the external loads.
To achieve this let's cut a tiny slice of the beam using two sections at $x$ and $x + \Delta{x}$.
We'll concider that the external forces consists of:
 - transverse distributed load with dencity $q(x)$ with two components - $q_y(x)$ and $q_z(x)$
 - axial distributed load with dencity $t(x)$
 - distributed moment with dencity $m(x)$ with it's three components $m_x(x)$, $m_y(x)$ and $m_z(x)$.

```python
Image(filename="Fig_4.jpg", width=400, height=200)
```

Since the two slices are very, very close, the dencity of the loads will not change significantly, so we can concider them as constants.\
If we substitude the distributed loads with their equivalent we'll get:


$$(3)\hspace{60pt} F = q(\xi_1)\Delta{x},\hspace{10pt} T = t(\xi_2)\Delta{x} $$

where $\xi_1$ and $\xi_2$ are some abscissas between $x$ and $x + \Delta{x}$:


$$x < \xi_1 < x + \Delta{x},\hspace{20pt} x < \xi_2 < x + \Delta{x}$$

The force $F$ is applied somewhere between $C$ and $C'$ (fig. 5,  $0 < \lambda < 1$)

```python
Image(filename="Fig_5.jpg", width=400, height=200)
```

The three equations of balance for the tiny cut element are:

$$(4)\hspace{20pt}\begin{align}&\sum{H} = 0\hspace{20pt} N(x + \Delta{x}) - N(x) + T = 0\\
&\sum{V} = 0\hspace{20pt} Q(x + \Delta{x}) - Q(x) + F = 0\\
&\sum{M} = 0\hspace{15pt} M(x + \Delta{x}) - M(x) - Q(x)\Delta{x} + F\lambda\Delta{x} = 0
\end{align}$$


If we substitude the equivalen loads from $(3)$ in the three equations of balance in $(4)$ we get:

$$(5)\hspace{50pt}\begin{align}&\frac{N(x + \Delta{x}) - N(x)}{\Delta{x}} = - t(\xi_2)\\\\
&\frac{Q(x + \Delta{x}) - Q(x)}{\Delta{x}} = - q(\xi_1)\\\\
&\frac{M(x + \Delta{x}) - M(x)}{\Delta{x}} = Q(x) + \lambda{q(\xi_1)}\Delta{x}
\end{align}$$

Let's assume that $\Delta{x}$ is tending to zero, in other words we move section $C'$ towards section $C$.\
The left parts of the equations $(5)$ will tend to the derivatives of the functions $N(x)$, $Q(x)$ and $M(x)$\
and the abscissas $\xi_1$ and $\xi_2$ will tend to $x$. Also in the third equation in $(5)$ as $\Delta{x}\rightarrow 0$, then $\lambda{q(\xi_1)}\Delta{x}\rightarrow 0$.\
Finally for the dependency ot the Internal forces and the external loads in a cross - section we get:

$$(6)\hspace{100pt}\begin{align}&\frac{dN}{dx} = -t(x)\\\\
&\frac{dQ}{dx} = - q(x)\\\\
&\frac{dM}{dx} = Q(x)
\end{align}$$
